# Langkah 2

Membaca dan menampilkan gambar dari variabel urls

````python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://avatars.githubusercontent.com/u/69801752?v=4"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
````

# Langkah 3

Melihat ukuran gambar

````python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
````

# Langkah 4

Mengakses pixel dan memberikan garis diagonal

````python
from ctypes import c_void_p
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3= cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

# Tugas 2

Membuat garis vertikal dan menyilang diagonal pada image keluaran

````python
# Garis vertikal
image_with_vertical_line = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_with_vertical_line[:, int((lebar/2))] = [255,255,255]

final_frame = cv.hconcat((image_2, image_with_vertical_line))
cv2_imshow(final_frame)
````

````python
# Garis silang
image_with_cross_line = np.array(cv.cvtColor(image, cv.COLOR_BGR2RGB))
image_with_cross_line[np.arange(len(image_with_cross_line)), np.arange(len(image_with_cross_line))] = [255,255,255]
image_with_cross_line[np.arange(len(image_with_cross_line)), np.flip(np.arange(len(image_with_cross_line)))] = [255,255,255]

final_frame = cv.hconcat((image_2, image_with_cross_line))
cv2_imshow(final_frame);
````

### Pertanyaan
Identifikasi kenapa menggunakan konversi warna BGR2RGB pada matplotlib dan tanpa konversi pada opencv

### Jawaban
Karena ketika membaca gambar dengan menggunakan OpenCV ```cv.imread()``` default warnanya adalah BGR jadi jika langsung ditampilkan maka warna dari gambar akan terlihat aneh karena urutan warna yang tidak sesuai. Jadi ketika gambar akan ditampilkan dengan ```plt.imshow()``` maka harus dikonversi terlebih dahulu dengan menggunakan ```cv.cvtColor("", cv.COLOR_BGR2RGB)```

````python
def draw(image, box_width, box_height, start, color):
  height = image.shape[0]
  width = image.shape[1]

  rectange_width = (int(width) - box_width) - 228
  rectange_height = (int(height) - box_height) - (start)

  image[start:-rectange_height, 228:-rectange_width] = color

  return image
````

````python
image = draw(ktp_img, 50, 12, 182,  [249, 115, 22])
image = draw(image, 100, 12, 200, [168, 85, 247])

plt.imshow(image)
plt.title("Kelompok 10 \n Sensor RT/RW dan Kel/Desa")
plt.axis("off")
````
